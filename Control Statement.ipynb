```python
# Control Statement
# break: break statement terminate the loop
# continue: continue statement skip the loop
```

```python
for i in range(20):
    if i==5:
        break
    else:
        print(i)
#         print(i,end =" ")      (end=" ") to print vertically
```

```python
for i in range(10):
    if i==5:
        continue
    else:  
        print(i,"Hello World")
```

```python
for i in range(1,21):
    if i%2!=0:
        continue
    print(i)
```

```python
# wap to detect the given number is prime or composite using while loop.
i=2
n=int(input("Enter the number:"))
if n>=2:
    while i<n//2+1:
        if n%i==0:
            print(n,"is composite")
            break
        i=i+1
    else:
        print(n,"is prime")
else:
    print(n,"The no should greater then 1")
```
