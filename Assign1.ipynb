```python
n=int(input("Enter the value of n:"))
all_total = 0
bill=str()
for i in range(n):
    name=input("Enter name:")
    price=int(input("Enter the price:"))
    quantity=int(input("Enter the quantity:"))
    total=price*quantity
    all_total = all_total + total
    bill=bill + name+" "+str(price)+" "+str(quantity) +" "+str(total)+"\n"
    
print(bill)
print(all_total)
    
```

```python
# Wap to create a result management systemof user
# name
# subjects(physics, chem)
# total
# percentage
# grade
```

```python
name=input("Enter the name of Student:")
s1=int(input("Enter the marks of physics:"))
s2=int(input("Enter the mark of chemistry:"))
t=s1+s2
p=t/2
print("Name=",name)
print("Total=",t)
print("Percentage=",p)

if p>=90:
    print("Grade=","A+")
elif p>=80 and p<=90:
    print("Grade=","A")
elif p>=70 and p<=80:
    print("Grade=","B+")
elif p>=60 and p<=70:
    print("Grade=","B")
elif p>=50 and p<=60:
    print("Grade=","C") 
elif p>=40 and p<=50:
    print("Grade=","D+")  
else:
    print("Grade=","Fail")
```

```python
name=input("Enter the name of Student:")
s1=int(input("Enter the marks of physics:"))
s2=int(input("Enter the mark of chemistry:"))
t=s1+s2
p=t/2
print("Name=",name)
print("Total=",t)
print("Percentage=",p)

if p>=80:
    print("Distinction")
elif p>=60 and p<=8
    print("First Division")
elif p>=50 and p<=60:
    print("Second Division")
elif p>=40 and p<=50:
    print("Third Division")
else:
    print("Fail")
```

```python
result = str()
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    physics = int(input("Enter physics = "))
    chemistry = int(input("Enter chemistry = "))
    math = int(input("Enter math = "))
    english = int(input("Enter english = "))
    nepali = int(input("Enter nepali = "))
    total = physics+chemistry+ math+english+nepali
    per = total/5
    if per >= 80:
        grade = 'A'
    elif per >= 60:
        grade = 'B'
    elif per >= 45:
        grade = 'C'
    else:
        grade = 'F'
    result = result + name + "\t"+str(physics) +"\t"+ str(chemistry)+"\t"+str(math)+"\t"+str(english)+"\t"+str(nepali)+"\t"+str(total)+"\t"+str(per)+"\t"+str(grade)+"\n"

print('Name\tphysics\tchemistry\tmath\tenglish\tnepali\ttotal\tper\tgrade')
print(result)
```
