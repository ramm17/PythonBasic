```python
#Factorial 5!=1*2*3*4*5
fac=1
i=1
n=int(input("Enter n="))
while i<=n:
    fac=fac*i
    i=i+1
print(fac)
```

```python
a="Hello world"
l=len(a)
i=0
while i<l:
    print(a[i])
    i=i+1
```

```python
a="Hello world"
l=len(a)
i=0
while i<l:
    if a[i]!=" ":
        print(a[i],end="")
    i=i+1
```

```python
#break and continue statement
i=0
n=10
while i<n:
    if i==5:
        break
    print(i)
    i=i+1
```

```python
a="Hello world"
l=len(a)
i=0
while i<l:
    if a[i]==" ":
        break
    print(a[i],end="")
    i=i+1
```

```python
a="Hello world"
l=len(a)
i=0
while i<l:
    if a[i]==" ":
        i=i+1
        continue
    print(a[i],end="")
    i=i+1
```
