```python
# Write a python program which will keeps adding a stream numbers given by the user. The adding stops as soon as the user press the q key.

sum=0
while(True):
    userInput=input("Enter the price or press 'q' to quit:\n")
    if (userInput!='q'):
        sum=sum+int(userInput)
        print(f"Your order total so far {sum}")
    else:
        print(f"Your total bill is {sum}. Thanks for Visiting us.")
        break
```
